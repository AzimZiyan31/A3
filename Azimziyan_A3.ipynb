```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, roc_auc_score, roc_curve, auc
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
# Load the dataset
loan_status = pd.read_csv("D:\\SCMA\\Data\\loan_data.csv.csv")
```

```python
# Display the dataset and its structure
print(loan_status.head())
print(loan_status.info())
print(loan_status.describe())
```

```python
# Plot missing values
sns.heatmap(loan_status.isnull(), cbar=False, cmap='viridis')
plt.show()
```

```python
# Replace missing values with the mean for numeric columns
loan_status.fillna(loan_status.mean(), inplace=True)
```

```python
# Checking missing values after filling it with mean values of the column 
missing_info = loan_status.isnull().sum()
print("Missing Values Information:\n", missing_info)
```

```python

# Correlation matrix and heatmap
cor_matrix = loan_status.corr()
print(cor_matrix)
sns.heatmap(cor_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Boxplot
sns.boxplot(x='loan_status', y='LoanAmount', hue='ApplicantIncome', data=loan_status)
plt.title("Boxplot of Age by Loan Status and ApplicantIncome")
plt.xlabel("Loan Status and ApplicantIncome")
plt.ylabel("LoanAmount")
plt.show()
```

```python
# Logistic Regression
# Assuming the unique values are 'Approved' and 'Not Approved'
# Convert them to 0 and 1
loan_status['loan_status'] = loan_status['loan_status'].apply(lambda x: 1 if x == 'Approved' else 0)
```

```python
X = loan_status.drop('loan_status', axis=1)
y = loan_status['loan_status']
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
pred_prob = model.predict_proba(X_test)[:, 1]
pred_class = (pred_prob >= 0.5).astype(int
```

```python
# Confusion Matrix
confusion = confusion_matrix(y_test, pred_class)
print("Confusion Matrix:\n", confusion)
```

```python
# ROC and AUC
roc_auc = roc_auc_score(y_test, pred_prob)
print(f"AUC-ROC: {roc_auc}")
```

```python
fpr, tpr, thresholds = roc_curve(y_test, pred_prob)
plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.2f}')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
# Decision Tree
model_tree = DecisionTreeClassifier(random_state=123)
model_tree.fit(X_train, y_train)
```

```python
pred_prob_tree = model_tree.predict_proba(X_test)[:, 1]
pred_class_tree = (pred_prob_tree >= 0.5).astype(int)
```

```python
confusion_tree = confusion_matrix(y_test, pred_class_tree)
print("Confusion Matrix (Decision Tree):\n", confusion_tree)
```

```python
roc_auc_tree = roc_auc_score(y_test, pred_prob_tree)
print(f"AUC-ROC (Decision Tree): {roc_auc_tree}")
```

```python
fpr_tree, tpr_tree, thresholds_tree = roc_curve(y_test, pred_prob_tree)
plt.plot(fpr_tree, tpr_tree, label=f'AUC = {roc_auc_tree:.2f}')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve (Decision Tree)')
plt.legend(loc='lower right')
plt.show()
```

```python
# Plot Decision Tree
plt.figure(figsize=(20,10))
plot_tree(model_tree, filled=True, feature_names=X.columns, class_names=['Not Approved', 'Approved'])
plt.show()
```
